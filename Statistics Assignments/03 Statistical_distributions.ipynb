1. By using distribution=np.random.name_distribution([],[]), write the name of distribution of your choice in place of `name_distribution`and fill out the bracket with your choice again. Then please, 

a) Draw the histogram and interpret

```python
import numpy as np
import matplotlib.pyplot as plt
sample_dataset=np.linspace(-50,50,100)
normally_distributed_data = np.random.normal(np.mean(sample_dataset),
                                         np.std(sample_dataset),100)
plt.hist(normally_distributed_data, bins = 40);
```

   b) Draw Q-Q plot and interpret

```python
from scipy import stats
import matplotlib.pyplot as plt

sample_dataset.sort()
normally_distributed_data.sort()

plt.figure(figsize=(10,4), dpi = 200)

plt.scatter(normally_distributed_data, sample_dataset)
#res = stats.probplot(sample_dataset, plot=plt)
plt.title("Q-Q Plot")
plt.plot([np.min(normally_distributed_data), np.max(normally_distributed_data)],
         [np.min(sample_dataset), np.max(sample_dataset)], c='red')
plt.show()
```
