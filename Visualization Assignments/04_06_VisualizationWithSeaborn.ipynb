1. Plot a line plot with seaborn for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns

data= pd.read_csv("E:/user/Notebooks/data/owid-covid-data.csv", parse_dates=["date"], low_memory=False)
```

```python
countries=data[data['location'].isin(['Spain', 'France', 'Germany', 'Italy'])]
countries=countries[countries['date']>'2020-04-01']
countries

spain=data[data['location']=='Spain']
italy=data[data['location']=='Italy']
france=data[data['location']=='France']
germany=data[data['location']=='Germany']

plt.figure(figsize=(12, 8))
plt.title('Total deaths of countries')
sns.lineplot(x = 'date', y = 'total_deaths', data=spain, lw=4, label = 'SPAIN')
sns.lineplot(x = 'date', y = 'total_deaths', data=italy, lw=4, label = 'ITALY')
sns.lineplot(x = 'date', y = 'total_deaths', data=france, lw=4, label = 'FRANCE')
sns.lineplot(x = 'date', y = 'total_deaths', data=germany, lw=4, label = 'GERMANY')
plt.legend(['SPAIN','ITALY','FRANCE','GERMANY'], loc = 4)
plt.show() 
```

2. Plot a bar plot with seaborn for average death number that compares continents.

```python
plt.figure(figsize=(12, 8))

data1=data.groupby(by=["continent"]).mean()
plt.title('Total Deaths')
sns.barplot(x=data1.index,y = data1['total_deaths'].values, data =data1)
plt.xticks(rotation = 90)
plt.show()
```

3. Plot a histogram for daily deaths for any country you choose. Make four subplots for different bin counts and `kde` arguments.

```python
plt.figure(figsize=(16, 12))

plt.subplot(2,2,1)
turkey=data[data['location']=='Turkey'].new_deaths

sns.distplot(turkey.dropna(), kde = False,bins=50, color = '#FF914D')
plt.title("kde = False,bins=50")

plt.subplot(2,2,2)
sns.distplot(turkey.dropna(), kde = False,bins=20, color = '#22B573')
plt.title("kde = False,bins=20")

plt.subplot(2,2,3)
sns.distplot(turkey.dropna(),bins=50, color = '#FF914D')
plt.title("kde = True,bins=50")

plt.subplot(2,2,4)
sns.distplot(turkey.dropna(),bins=20, color = '#22B573')
plt.title("kde = True,bins=20")

plt.show()
```

4. Create a figure and three subplots containing boxplot, violin plot and swarm plot for daily deaths of two countries you choose.

```python
data2=data[data['location'].isin([ 'France', 'Germany']).dropna()]
plt.figure(figsize=(15, 5))
plt.subplot(1,3,1)
plt.title("box plot")
sns.boxplot(x ='location' ,  y = 'new_deaths', data =data2, palette = ['#FF914D', '#22B573'])


plt.subplot(1,3,2)
plt.title("violin plot")
sns.violinplot(x ='location' ,  y = 'new_deaths', data =data2, palette = ['#22B573', '#FF914D'])

plt.subplot(1,3,3)
plt.title("swarm plot")
sns.swarmplot(x ='location' ,  y = 'new_deaths', data =data2);
```
