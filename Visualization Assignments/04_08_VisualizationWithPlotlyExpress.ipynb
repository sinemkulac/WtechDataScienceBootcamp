```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt

data= pd.read_csv('E:/user/Notebooks/data/owid-covid-data.csv', parse_dates=["date"], low_memory=False)
```

1. Plot a line plot with plotly express for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.

```python
countries=data[data['location'].isin(['Spain', 'France', 'Germany', 'Italy'])]
countries=countries[countries['date']>'2020-04-01']

fig = px.line(countries, x="date", y="total_deaths", color='location')
fig.show()
```

2. Plot a bar plot with plotly express for average death number that compares continents.

```python
plt.figure(figsize=(12, 8))
data1=data.groupby(by=["continent"]).mean()
fig = px.bar(data_frame=data1, 
             x=data1.index, 
            y=data1['total_deaths'].values)
fig.show()
```

3. Plot a histogram for daily deaths for any country you choose. 

```python
turkey_data=data[(data['location'] =='Turkey')]
turkey_data

fig = px.histogram(turkey_data, y="new_deaths",x='date')
fig.show()
```
