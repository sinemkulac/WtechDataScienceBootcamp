1. Plot a bar chart for total cases of the 20 countries that havebiggest numbers.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data= pd.read_csv("E:/user/Notebooks/data/owid-covid-data.csv")

data1=data.groupby(by=["location"]).mean()
country_data=data1['total_cases'].nlargest(20)
plt.figure(figsize=(20, 10))
pd.options.display.float_format="{:.2f}".format
plt.bar(country_data.index, country_data, color = "#FF914D")
```

2. Plot a histogram for daily deaths for any country you choose. Make three subplots for different bins.

```python
turkey_data=data[(data['location'] =='Turkey') ]

turkey_data=turkey_data[(turkey_data['date']>'2020-06-04')]

plt.figure(figsize=(15, 4))
plt.subplot(1, 3, 1)
plt.hist(turkey_data['new_deaths'], color = "#22B573", orientation = 'vertical')
```

3. Plot a scatter plot of new cases and new death for Germany and France.

```python
#data_=data[data['location'].isin(['Germany', 'France'])]
plt.figure(figsize=(10, 6))
plt.title('New cases & new deaths')
for location in ['Germany', 'France']:
    plt.scatter(data[data.location == location].new_cases, 
                data[data.location == location].new_deaths, label = location)
#plt.scatter(data_.new_cases, data_.new_deaths, c = '#FF914D')
plt.xlabel('new cases')
plt.ylabel('new deaths')
plt.xticks(rotation = 90, fontsize = 9)
plt.legend(['GERMANY','FRANCE'], loc = 4)
plt.show()
```

4. Plot a boxplot for daily deaths for any country you choose.

```python
plt.figure(figsize=(10, 6))
plt.boxplot(data[data.location == 'Turkey'].new_deaths.dropna())
plt.show()
```

5. Calculate the total case for each continent and plot a pie chart 

```python
plt.figure(figsize=(10, 6))

groupby_loc_data=data.groupby(by=["location"]).mean()
groupby_loc_data=groupby_loc_data['total_cases'].nlargest(20)

groupby_loc_data['World']=0


plt.pie(groupby_loc_data, labels=groupby_loc_data.index, autopct='%1.1f%%',shadow=True, startangle=90)
plt.show()
```
