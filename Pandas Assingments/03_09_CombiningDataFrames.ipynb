````python
#data
pd.merge(data,data2,left_on='id',right_on='id',how='right')
````

````python
pd.merge(data,data2,left_on='id',right_on='id',how='left')
````

````python
pd.merge(data,data2,left_on='id',right_on='id',how='outer')
````
