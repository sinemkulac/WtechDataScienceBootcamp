```python
def make_categorize(imdb_score):
    if imdb_score >= 7.5:
        return 'good'
    else: 
        return 'bad'
    
```

```python
movie['imdb_score_group'] = movie['imdb_score'].apply(make_categorize)
```

```python
movie.groupby('imdb_score_group')["imdb_score"].count()
```

3.Is it possible to convert categorical variable into numerical one? If so, how can you make that?

```python
def make_numerical(color):
    if color=="Black and White":
        return 0
    else: 
        return 1
```

```python
movie["color"]=movie["color"].apply(make_numerical)
```
